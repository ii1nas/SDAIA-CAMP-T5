 1. Expense and Income Entry

Users can input their expenses and income, including the amount,category, and date.



```python
# Initialize an empty dictionary to store transactions
transactions = {}

# Function to add a transaction
def add_trans():
    global transactions
    date = input("Enter the date (YYYY-MM-DD): ")
    category = input("Enter the category: ")
    amount = float(input("Enter the amount: "))


    transaction = {"date": date, "category": category,"amount": amount}
    transactions[date] = transaction

# Function to view all transactions
def view_trans():
    for key, value in transactions.items():
        print(f"Date: {value['date']}, Amount: {value['amount']}, Category: {value['category']}")
# Main loop for the program
while True:
    print("\n1. Add Transaction")
    print("2. Exit")

    choice = input("\nEnter your choice: ")

    if choice == '1':
        add_trans()
    elif choice == '2':
        break
    else:
        print("Invalid choice. Please try again.")
```

#### 3. View Transactions

Users can view a list of their transactions, sorted by date or category. They can also update and delete their own transactions.

```python
transactions = {}

def add_trans(date, amount, category):
    if category in transactions:
        transactions[category].append((date, amount))
    else:
        transactions[category] = [(date, amount)]

def view_trans(sort_by='date'):
    for category, details in transactions.items():
        print(f"{category}:")
        if sort_by == 'date':
            details.sort(key=lambda x: x[0])
        for date, amount in details:
            print(f"Date: {date}, Amount: {amount}")

def update_trans(category, index, new_amount):
    if category in transactions and index < len(transactions[category]):
        transactions[category][index] = (transactions[category][index][0], new_amount)
    else:
        print("Transaction not found.")

def delete_trans(category, index):
    if category in transactions and index < len(transactions[category]):
        del transactions[category][index]
    else:
        print("Transaction not found.")

# Example transactions
add_trans('2022-03-25', 10, 'food')
add_trans('2022-03-5', 50, 'transportation')
add_trans('2022-03-15', 500, 'salary')
add_trans('2022-03-10', 5, 'food')

# Display transactions sorted by date
view_trans()

# Update a transaction
update_trans('food', 0, 25)

# Display transactions sorted by category
view_trans(sort_by='category')

# Delete a transaction
delete_trans('food', 0)

# Display transactions sorted by date
view_trans()
```

 4. Basic Reports

Generate basic reports such as:

-   Total expenses
-   Total income
-   Spending by category
-   Transactions in each category
-   Average transaction value by category

```python
transactions = {
    'food': [('2022-03-25', 10), ('2022-03-10', 5)],
    'transportation': [('2022-03-5', 50)],
    'salary': [('2022-03-15', 500)]
}

def total_expenses():
    total_expense = sum(sum(amount for _, amount in details) for details in transactions.values())
    return total_expense

def total_income():
    total_income = sum(sum(amount for _, amount in details) for category, details in transactions.items() if category == 'salary')
    return total_income

def spending_by_category():
    category_spending = {category: sum(amount for _, amount in details) for category, details in transactions.items()}
    return category_spending

def transactions_by_category():
    return {category: len(details) for category, details in transactions.items()}

def avg_transaction_value_by_category():
    avg_values = {category: sum(amount for _, amount in details) / len(details) for category, details in transactions.items()}
    return avg_values

# Basic Reports
print(f"Total Expenses: {total_expenses()}")
print(f"Total Income: {total_income()}")
print("Spending by Category:")
for category, spending in spending_by_category().items():
    print(f"{category}: {spending}")
print("Transactions in Each Category:")
for category, count in transactions_by_category().items():
    print(f"{category}: {count}")
print("Average Transaction Value by Category:")
for category, avg_value in avg_transaction_value_by_category().items():
    print(f"{category}: {avg_value}")
```
