The purpose of an activation function is to introduce non-linear transformations to the networkâ€™s computations. Without activation functions, the network would be limited to performing only linear transformations

## Sigmoid Function:

The sigmoid activation function, also referred to as the logistic function, is a widely applied mathematical function in machine learning. It transforms input values into a range spanning from 0 to 1.

![Sigmoid](https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/1200px-Logistic-curve.svg.png)


Characterized by an S-shaped curve, the sigmoid function is mathematically defined as f(x) = 1 / (1 + e^(-x)), where f(x) signifies the sigmoid output for a given input x, and e represents the mathematical constant approximately equal to 2.71828.

Commonly utilized in binary classification tasks, the sigmoid activation function generates probabilities indicating the likelihood of an input belonging to a specific class. With its differentiable nature and output range, it proves beneficial for modeling nonlinear relationships and mapping inputs to probabilities

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))

sigmoid(-5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sigmoid():
    x = np.linspace(-10, 10, 100)  # Generate 100 equally spaced values from -10 to 10
    y = 1 / (1 + np.exp(-x))  # Compute the sigmoid function values

    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('Sigmoid Output')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()

plot_sigmoid()
```

## SoftMax

The SoftMax function is like a voting system. It takes in the outputs of all the neurons in a layer and decides which neuron should have the most influence on the final output. It does this by converting the outputs into probabilities, with the highest probability representing the most influence.

![Softax](https://miro.medium.com/v2/resize:fit:1400/1*ReYpdIZ3ZSAPb2W8cJpkBg.jpeg)


```python
def softmax(x):
  """
  This function computes the softmax activation function for a given input vector.

  Args:
      x: A numpy array representing the input vector.

  Returns:
      A numpy array representing the softmax output vector.
  """
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y = exp_x / sum_exp_x
  return y

# Example usage
x = np.array([1, 2, 3])
y = softmax(x)
print(y)

```

The softmax function outputs a probability distribution over multiple classes, so it is commonly used in the context of classification tasks. However, it doesnâ€™t make sense to plot the softmax function itself since it takes a vector of values as input and returns a vector of probabilities.

```python
def plot_softmax():
    # Define the input vector
    x = np.array([0.3, 2.9, 4.0])

    # Calculate the softmax probabilities
    y = softmax(x)

    # Generate the plot
    plt.bar(range(len(x)), y)
    plt.xlabel('Input Index')
    plt.ylabel('Softmax Probability')
    plt.title('Softmax Function Output')
    plt.show()

plot_softmax()
```

## Tanh Function

The hyperbolic tangent (tanh) activation function is frequently employed in neural networks. Serving as an extension of the sigmoid function, it maps input values to a range spanning from -1 to 1.

With its symmetric nature around the origin, the tanh activation function yields outputs in both positive and negative realms. This feature proves advantageous for capturing nonlinear connections, rendering it a common choice in recurrent neural networks (RNNs) and within the hidden layers of feedforward neural networks

```python
#Creating Tanh Function
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

#Testing
tanh(50)
```

```python
# function to draw tanh function

import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def plot_tanh():
  x = np.linspace(-10, 10, 100)
  y = tanh(x)
  plt.plot(x, y)
  plt.xlabel('Input')
  plt.ylabel('Tanh Output')
  plt.title('Tanh Activation Function')
  plt.grid(True)
  plt.show()

plot_tanh()
```

## The Rectified Linear Unit (ReLU)

The Rectified Linear Unit (ReLU) activation function is a prevalent mathematical operation in neural networks. It employs a straightforward thresholding mechanism on input values, converting negative values to zero while preserving positive values.

Expressed mathematically as f(x) = max(0, x), where f(x) denotes the ReLU output for a given input x.

Renowned for its simplicity and computational efficacy, ReLU has gained widespread adoption, particularly in deep learning frameworks. Its capacity to mitigate the vanishing gradient issue and expedite convergence during training makes it a favored choice in various architectures.

```python
def relu(x):
    """
    This function computes the rectified linear unit (ReLU) activation function for a given input vector.

    Args:
        x: A numpy array representing the input vector.

    Returns:
        A numpy array representing the ReLU output vector.
    """
    return np.maximum(0, x)

# Example usage
x = np.array([-1, 0, 1, 2, 3])
y = relu(x)
print(y)


```

```python
# Plot the ReLU function
def plot_relu():
    x = np.linspace(-5, 5, 100)
    y = relu(x)
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('ReLU Output')
    plt.title('ReLU Activation Function')
    plt.grid(True)
    plt.show()

plot_relu()
```


## Exponential Rectified Linear Unit (ERELU)

It is a variant of the popular ReLU (Rectified Linear Unit) activation function used in neural networks. Like ReLU, ERELU promotes faster convergence and reduces the likelihood of the vanishing gradient problem. The difference is that ERELU adds an exponential term to negative inputs, which provides a soft, non-zero gradient when the input is less than zero, unlike ReLU which outputs zero for negative inputs.

Useful when dealing with deep learning models where ReLU's zero output for negative inputs might cause issues with learning, especially in deeper layers of the network.

ERELU can help in maintaining a flow of gradients, thus aiding the training process. It is particularly useful in scenarios where learning from negative inputs is crucial and where avoiding dead neurons (neurons that only output zero) is essential.

```python
def ereelu(x):
    """
    This function computes the exponential rectified linear unit (ERELU) activation function for a given input vector.

    Args:
        x: A numpy array representing the input vector.

    Returns:
        A numpy array representing the ERELU output vector.
    """
    return np.where(x >= 0, x, np.exp(x) - 1)

# Plot the ERELU function
def plot_ereelu():
    x = np.linspace(-5, 5, 100)
    y = ereelu(x)
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('ERELU Output')
    plt.title('ERELU Activation Function')
    plt.grid(True)
    plt.show()

plot_ereelu()
```

In summary, the choice of activation function plays a crucial role in the performance and efficiency of neural networks. Each activation function offers unique characteristics tailored to specific tasks and architectures. While sigmoid and tanh are suitable for certain types of mappings and binary classification, ReLU and its variants excel in deep learning scenarios, overcoming issues like vanishing gradients. Softmax stands out for multi-class classification, providing a robust probability distribution across multiple classes. Understanding these distinctions enables practitioners to optimize neural network designs for diverse applications.

| Activation Function | When to Use                                    |
|---------------------|------------------------------------------------|
| Sigmoid             | Useful for binary classification problems where you need probabilities as outputs; often used in the output layer for binary outcomes. |
| Softmax             | Ideal for multi-class classification problems; typically used in the output layer to represent probabilities of each class. |
| ReLU                | Commonly used in hidden layers of neural networks; helps with fast training and addressing the vanishing gradient problem for positive inputs. |
| EReLU               | Suitable for deeper networks where ReLU might lead to dead neurons; helps maintain gradient flow and is useful when learning from negative values is crucial. |

This table guides you on choosing the appropriate activation function based on the specific requirements of your neural network architecture and the nature of your problem.
